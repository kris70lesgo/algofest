```python
import pandas as pd
import plotly.graph_objects as go

# Updated path - go up one level, then into src/data/raw
aapl = pd.read_csv('../src/data/raw/AAPL.csv', index_col=0, parse_dates=True)

# Rest of your code remains the same
print("Data shape:", aapl.shape)
print("\nColumns:", aapl.columns.tolist())
print("\nLast 5 days:")
print(aapl.tail())

# Plot price chart
fig = go.Figure()
fig.add_trace(go.Scatter(x=aapl.index, y=aapl['Close'], name='AAPL Close'))
fig.update_layout(title='AAPL Price Chart', xaxis_title='Date', yaxis_title='Price')
fig.show()
```
